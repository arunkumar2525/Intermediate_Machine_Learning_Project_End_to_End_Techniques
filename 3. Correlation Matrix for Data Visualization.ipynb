```python
# correlation gives the relation b/n the changes are in between two variable
# If both variable changes in same direction then it is +ve or else vise versa
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
fileName='pima-indians-diabetes.csv'
names=['perg','plas','pres','skin','test','mass','pedi','age','class']
dataset=pd.read_csv(fileName,names=names)
print(dataset.head(1))
correlation=dataset.corr()
# plot correlation matrix
fig=plt.figure()
ax=fig.add_subplot(111)
cax=ax.matshow(correlation,vmin=-1,vmax=1)
fig.colorbar(cax)
ticks=np.arange(0,9,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names)
ax.set_yticklabels(names)
plt.show()
```
