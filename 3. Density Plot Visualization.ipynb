```python
# It is another way of getting a quick idea of the distribution of each attribute(column)
import pandas as pd
import matplotlib.pyplot as plt
fileName='pima-indians-diabetes.csv'
names=['perg','plas','pres','skin','test','mass','pedi','age','class']
dataset=pd.read_csv(fileName,names=names)
print(dataset.head(5))
dataset.plot(kind='density',layout=(3,3),sharex=False)
dataset.plot(kind='density',subplots=True,layout=(3,3),sharex=False)
plt.show()
```
